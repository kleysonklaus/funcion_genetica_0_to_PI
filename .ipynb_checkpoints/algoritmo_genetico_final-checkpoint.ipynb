```python
import random as rn
def p_mutacion (probabilidad):
    num = rn.randint(1,101)
    if (num <=probabilidad):
        num = num /100
        return True
    else:
        return False 
def p_cruce (probabilidad):
    num = rn.randint(1,100)
    if (num <= probabilidad):
        num = num /100
        return True
    else:
        return False
def poblacion_random(x):
    a = []
    for i in range (x):
        a.append(rn.randint(0,1))
    return a

def bin_to_decimal(x):
    d1 = 2
    d2 = 64
    entero = 0
    decimal = 0
    for i in range(len(x)):
        if (i<2):
            entero = entero + x[i]*d1
            d1 = d1/2
        else:
            decimal = decimal + x[i]*d2
            d2 = d2/2
    return round(entero), round(decimal)

def imprimir_poblacion(mundo):
    for i in range (len(mundo)):
        print (i+1,": ",mundo[i]," valor = ",bin_to_decimal(mundo[i]))
 
#from decimal import Decimal

def extraer_datos(mundo , _print):
    datos = []
    entero = []
    decimal = []
    for i in range(len(mundo)):
        ent,dec = bin_to_decimal(mundo[i])
        entero.append(ent)
        decimal.append(dec)
        
    for i in range (len(entero)):
        entero[i] = float(entero[i])
        if ( decimal[i] < 10 ):
            decimal[i] = decimal[i] * 0.1
        elif ( decimal[i] < 100 ):
            decimal[i] = decimal[i] * 0.01
        else:
            decimal[i] = decimal[i] * 0.001
        decimal[i] = round(decimal[i],3)
        datos.append( entero[i]+decimal[i] )
        if _print:
            print ( datos[i] )
    #retorna un array con los datos decimales
    return datos

def seleccion_cruce(mundo , _print):
    nuevo_mundo = []
    valores = extraer_datos(mundo , False)
    ordenado = sorted(valores)
    cantidad = rn.randint(2,len(ordenado))
    crucez = []
    valor = 0
    for i in range (len (valores)):
        for j in range (cantidad):
            if ( valores[i] == ordenado[j] ):
                valor = 1
                break;
            else:
                valor = 0
        crucez.append(valor)
        if( valor == 1 ):
            nuevo_mundo.append( mundo[i] )
    #print (crucez)
    if _print:
        print ("Cantidad de Seleccionados: ", cantidad)
        for i in range (len(valores)):
            if ( crucez[i] == 1 ):
                print (mundo[i]," = ",valores[i], "\tSeleccionado!" )
            else:
                print (mundo[i]," = ",valores[i], "\tMUERE" )
        print ("array de indices: ",crucez)
    return nuevo_mundo,crucez,cantidad,valores

def mutar_poblacion (mundo,array_de_cruce,valores,pro_indi,_print):
    mundo_mutado = []
    x_temporal=[]
    for i in range ( len(mundo) ):
        x_temporal = mundo[i]
        if ( array_de_cruce[i] == 1 ):
            posibilidad = p_mutacion (pro_indi)
            if posibilidad:
                indice_random = rn.randint(0,len(x_temporal)-1)
                if _print:
                    print (mundo[i]," = ",valores[i], "\tVIVO \t- MUTADO!")
                    print ("Antes = ",mundo[i]," = ","Antiguo valor",valores[i])
                    print ("indice a cambiar: ",indice_random+1)
                    
                if x_temporal[indice_random] == 1:
                    x_temporal[indice_random] = 0
                else:
                    x_temporal[indice_random] = 1
                if _print:
                    e,d = bin_to_decimal(mundo[i])
                    print ("Nuevo = ",mundo[i]," = Nuevo valor = ",e,",",d)
                #mundo_mutado.append ( mundo[i] )

            elif _print:
                print (mundo[i]," = ",valores[i], "\tVIVO \t- No Mutado")
                
            mundo_mutado.append ( mundo[i] )
        else:
            if _print:
                print (mundo[i]," = ",valores[i], "\tMUERTO")
    return mundo_mutado

def reproducir(padre1,padre2):
    arr1 = []
    arr2 = []
    desde = rn.randint(2,len(padre1)-1)
    
    for i in range ( len(padre1) ):
        if i < desde:
            arr1.append( padre1[i] )
        else:
            arr1.append( padre2[i] )
    desde2 = rn.randint(2,len(padre2)-1)
    for i in range ( len(padre2) ):
        if i < desde2:
            arr2.append( padre2[i] )
        else:
            arr2.append( padre1[i] )
    return arr1,arr2,desde,desde2
        
def cruzar_poblacion(mundo,probabilidad,_print):
    nuevo_mundo = []
    tamano_mundo = len(mundo)
    if tamano_mundo % 2== 0:
        tamano_mundo = int (tamano_mundo/2)
        print ("numero de cruces: ",tamano_mundo)
        indice = 0
        for i in range (tamano_mundo):
            pro = p_cruce(probabilidad)
            if pro:
                #hijos, pivote desde bits
                hijo1,hijo2,p_h1,p_h2, = reproducir ( mundo[indice] , mundo[indice+1],)
                
                nuevo_mundo.append( hijo1 )
                nuevo_mundo.append( hijo2 )
                
                if _print:
                    print ("cruce: ",i+1)
                    print ("padres= ",mundo[indice]," - ",mundo[indice+1]," ",probabilidad,"% SI se reproduce")
                    print ("hijos : ",hijo1," - ",hijo2," ->",p_h1,"-",p_h2)
            else:
                nuevo_mundo.append( mundo[indice] )
                nuevo_mundo.append( mundo[indice+1] )
                
                if _print:
                    print ("cruce: ",i+1)
                    print ("padres= ",mundo[indice]," - ",mundo[indice+1]," ",probabilidad,"% NO se reproduce")
                    print ("hijos : ",mundo[indice]," - ",mundo[indice+1]," = se mantienen")
            indice = indice +2
            
    else:
        tamano_mundo = tamano_mundo -1
        tamano_mundo = int (tamano_mundo/2)
        print ("numero de cruces: ",tamano_mundo+1)
        indice = 0
        for i in range (tamano_mundo):
            pro = p_cruce(probabilidad)
            if pro:
                #hijos, pivote desde bits
                hijo1,hijo2,p_h1,p_h2, = reproducir ( mundo[indice] , mundo[indice+1])
                
                nuevo_mundo.append( hijo1 )
                nuevo_mundo.append( hijo2 )
                
                if _print:
                    print ("cruce: ",i+1)
                    print ("padres= ",mundo[indice]," - ",mundo[indice+1]," ",probabilidad,"% SI se reproduce")
                    print ("hijos : ",hijo1," - ",hijo2," ->",p_h1,"-",p_h2)
            else:
                nuevo_mundo.append( mundo[indice] )
                nuevo_mundo.append( mundo[indice+1] )
                
                if _print:
                    print ("cruce: ",i+1)
                    print ("padres= ",mundo[indice]," - ",mundo[indice+1]," ",probabilidad,"% NO se reproduce")
                    print ("hijos : ",mundo[indice]," - ",mundo[indice+1]," = se mantienen")
            indice = indice +2
            
        #para el que esta solo
        pro = p_cruce(probabilidad)
        if pro:
            #hijos, pivote desde bits
            hijo1,hijo2,p_h1,p_h2, = reproducir ( mundo[indice] , mundo[0])
            nuevo_mundo.append( hijo1 )
            nuevo_mundo.append( hijo2 )
            
            if _print:
                print ("cruce: ",i+2)
                print ("padres= ",mundo[indice]," - ",mundo[0]," ",probabilidad,"% SI se reproduce")
                print ("hijos : ",hijo1," - ",hijo2," ->",p_h1,"-",p_h2)
        else:
            nuevo_mundo.append( mundo[indice] )
            nuevo_mundo.append( mundo[0] )
            
            if _print:
                print ("cruce: ",i+2)
                print ("padres= ",mundo[indice]," - ",mundo[0]," ",probabilidad,"% NO se reproduce")
                print ("hijos : ",mundo[indice]," - ",mundo[0]," = se mantienen")
            
            
    return nuevo_mundo
```
